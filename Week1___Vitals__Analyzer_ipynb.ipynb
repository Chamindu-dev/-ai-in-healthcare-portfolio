```python
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/DeegayuA/Teaching-data/main/patient_vitals.csv'
df=pd.read_csv(url)
```

```python
print("--first 5 rows--")
print(df.head())
```

```python
print("\n--data information summary--")
df.info
```

```python
average_age=df['Age'].mean()
print(f"Report Insight 01:The average patient age is{average_age:.2f} years.")
```

```python
max_heart_rate=df['HeartRate'].max()
min_heart_rate=df['HeartRate'].min()
print(f"Report Insight 02:The heart rate ranged from {min_heart_rate} bpm to {max_heart_rate} bpm.")
```

```python
hypertensive_patient=df[df['SystolicBP']> 140]
num_hypertensive=len(hypertensive_patient)
print(f"Report Insight 03:There are {num_hypertensive} patients with hypertensive crisis.")
```

```python
high_risk_female=df[(df['Gender']== 'Female') & (df['Age']> 60)]
print("\n---High Risk Group Report (Females over 60)---")
print(high_risk_female)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', bins=10, kde=True, color='skyblue')
plt.title('Distribution of Patient Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
